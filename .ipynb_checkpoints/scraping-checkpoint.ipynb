```python
import requests
import csv
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://movies.yahoo.co.jp/movie/')
soup = BeautifulSoup(r.content, "html.parser")

tags = soup.find_all("li",{"class":"col"})
with open('url.csv', 'w') as f:
    for tag in tags:
        url="https://movies.yahoo.co.jp/movie/"+tag.get("data-cinema-id")+"/"
        print(url)
        writer = csv.writer(f)
        writer.writerow([url])
```

```python
for i in range(2,11):
    r = requests.get("https://movies.yahoo.co.jp/movie/?page="+str(i))
    soup = BeautifulSoup(r.content, "html.parser")

    tags = soup.find_all("li",{"class":"col"})
    with open('url.csv', 'a') as f:
        for tag in tags:
            url="https://movies.yahoo.co.jp/movie/"+tag.get("data-cinema-id")+"/"
            print(url)
            writer = csv.writer(f)
            writer.writerow([url])
```
